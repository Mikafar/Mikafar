```python
# task 1
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.pns.hk/zh-hk/search?text=%E6%B2%B9&useDefaultSearch=false"
header = {
    "user-agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.82 Safari/537.36',
}
response = requests.get(url, headers=header)
http_string = response.text
html = BeautifulSoup(http_string, 'html.parser')
product_list = html.select(".productContainer")

name = html.select_one(".productInfo > .productName").text
unit = html.select_one(".productInfo > .productUnit").text
price = html.select_one(".productInfo .productPrice .currentPrice").text
img = html.select_one(".productImage img")['src']

dl = []
for item in product_list:
  name = item.select_one(".productInfo > .productName").text
  unit = item.select_one(".productInfo > .productUnit").text
  price = item.select_one(".productInfo .productPrice .currentPrice").text
  img = item.select_one(".productImage img")['src']
  dl.append([name, unit, price, img])

df = pd.DataFrame(dl, columns=['name', 'unit', 'price', 'img'])
df
```

```python
# task 2 hk.centanet
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://hk.centanet.com/findproperty/list/buy?q=Ha4NHoLyzEqBLZ4Wnt01FQ"
header = {
    "user-agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.82 Safari/537.36',
}
response = requests.get(url, headers=header)
http_string = response.text
html = BeautifulSoup(http_string, 'html.parser')
```

```python

infoList = html.select('.list')
dataList = []
for item in infoList:
  name = item.select_one('.title > .title-lg').text
  floor = item.select_one('.title > .title-sm').text
  address = item.select_one('.col-detail > .adress.tag-adress').text
  price = item.select_one('.price > span:nth-of-type(3)').text
  nSize = item.select_one('.area-block.usable-area > .num > span').text.replace(' ', '')
  size = item.select_one('.area-block.construction-area > .num > span').text
  info = html.select_one('.adress.floor-info > span').text.replace('\n', '').replace('\r', '').replace(' ', '')
  # age = item.select_one('.adress.floor-info > span > span').text
  # direction = item.select_one('.adress.floor-info > span > span:nth-of-type(3)').text
  # try:
  #  usage = item.select_one('.adress.floor-info > span > span:nth-of-type(5)').text.replace(' ', '').replace('\n', '')
  # except:
  #   usage = ''
  dataList.append([name, floor, address, price, nSize, size, info])

df = pd.DataFrame(dataList, columns=['name', 'floor', 'address', 'price', 'nSize', 'size', 'info'])
df.to_csv("dataList", index=False)
df
```

```python
# task 3
import requests
import pandas as pd

url = 'https://hk.centanet.com/findproperty/api/Post/Search'
header = {
    "user-agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.82 Safari/537.36',
    "Content-Type": "application/json; charset=utf-8",
}
data = {
    "bigestAndEstate": ["3-XSHNIHSXHT"],
    "displayTextStyle": "WebResultList",
    "offset": 0,
    "order": "Ascending",
    "pageSource": "search",
    "phaseAndEstate": [],
    "postType": "Sale",
    "size": 24,
    "sort": "Ranking"
}

response = requests.post(url, headers=header, json=data)
count = response.json()['count']
print(count)
dataList = response.json()['data']
i = 24

while True:
  if i >= 356:
    break
  data = {
    "bigestAndEstate": ["3-XSHNIHSXHT"],
    "displayTextStyle": "WebResultList",
    "offset": i,
    "order": "Ascending",
    "pageSource": "search",
    "phaseAndEstate": [],
    "postType": "Sale",
    "size": 24,
    "sort": "Ranking"
  }
  response = requests.post(url, headers=header, json=data)
  if response.status_code == 200:
    dataList.extend(response.json()['data'])
    i = i+24


# print(len(dataList))
# print(dataList[0]['id'])


```

```python

# the for loop can be also implemented in the while loop but for better understanding,
# I placed it outside the loop
infoList = []
for item in dataList:
  name = item['bigEstateName']

  try:
    size = item['areaInfo']['size']
  except:
    # print(item)
    size=""

  try:
    nSize = item['areaInfo']['nSize']
  except:
    # print(item)
    nSize=""

  address = item['address']
  price = item['priceInfo']['price']

  try:
    buildingGroup = item['bldgGrp']
    buildingGroupName = buildingGroup['bldgGrpName']
    url = buildingGroup['url']
  except:
    buildingGroupName = ""
    url = ""

  infoList.append([name, size, nSize, address, buildingGroupName, url])

df = pd.DataFrame(infoList, columns=['name', 'size', 'nSize', 'address', 'GroupName', 'url'])
df.to_csv
df
```

```python
# task 4 parknshop http request
# additional tasks
import requests
import pandas as pd

i = 1
dl = []
while True:
  url = f"https://api.pns.hk/api/v2/pnshk/products/search?fields=FULL&query=%E6%B2%B9&pageSize=18&currentPage={i}&sort=mostRelevant&useDefaultSearch=false&lang=zh_HK&curr=HKD"
  header = {
      "user-agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.82 Safari/537.36',
  }

  response = requests.get(url, headers=header)
  if response.status_code == 200:
    pages = response.json()['pagination']['pageSize']
    product = response.json()['products']
    for item in product:
      name = item['elabProductName']
      price = item['elabPrice']['value']
      size = item['contentSizeUnit']
      dl.append([name, price, size])

    i = i + 1
  if i > pages:
    break

df = pd.DataFrame(dl, columns=['name', 'price', 'size'])
df
```
